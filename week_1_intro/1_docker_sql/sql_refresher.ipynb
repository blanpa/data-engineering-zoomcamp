Joining the trip db with zone lookup db

```python
%%sql

SELECT 
    tpep_pickup_datetime,
    tpep_dropoff_datetime,
    tolls_amount ,
    CONCAT(zpu."Borough", ' / ', zpu."Zone") AS "pickup_loc",
    CONCAT(zdo."Borough", ' / ', zdo."Zone") AS "dropoff_loc"
FROM
    yellow_taxi_data t,
    zones zpu,
    zones zdo
where
	t."PULocationID" = zpu."LocationID" and
	t."DOLocationID" = zdo."LocationID"
	
limit 100

    

```

```python
%%sql

SELECT 
    tpep_pickup_datetime,
    tpep_dropoff_datetime,
    tolls_amount ,
    CONCAT(zpu."Borough", ' / ', zpu."Zone") AS "pickup_loc",
    CONCAT(zpu."Borough", ' / ', zpu."Zone") AS "dropoff_loc"
FROM
    yellow_taxi_data t,
    zones zpu
where
	t."PULocationID" = zpu."LocationID" and
	t."DOLocationID" = zpu."LocationID"
	
limit 100
```

```python
%%sql

SELECT 
    tpep_pickup_datetime,
    tpep_dropoff_datetime,
    tolls_amount ,
    CONCAT(zpu."Borough", ' / ', zpu."Zone") AS "pickup_loc",
    CONCAT(zdo."Borough", ' / ', zdo."Zone") AS "dropoff_loc"
FROM
    yellow_taxi_data t join zones zpu
    	on t."PULocationID" = zpu."LocationID"
    join zones zdo
    	on t."DOLocationID" = zdo."LocationID"	
limit 100;
```

LEFT, RIGHT and OUTER JOINs

```python
%%sql

SELECT 
    tpep_pickup_datetime,
    tpep_dropoff_datetime,
    tolls_amount ,
    CONCAT(zpu."Borough", ' / ', zpu."Zone") AS "pickup_loc",
    CONCAT(zdo."Borough", ' / ', zdo."Zone") AS "dropoff_loc"
FROM
    yellow_taxi_data t LEFT JOIN zones zpu
    	on t."PULocationID" = zpu."LocationID"
    join zones zdo
    	on t."DOLocationID" = zdo."LocationID"	
limit 100;
```

```python
%%sql

SELECT 
    tpep_pickup_datetime,
    tpep_dropoff_datetime,
    tolls_amount ,
    CONCAT(zpu."Borough", ' / ', zpu."Zone") AS "pickup_loc",
    CONCAT(zdo."Borough", ' / ', zdo."Zone") AS "dropoff_loc"
FROM
    yellow_taxi_data t RIGHT JOIN zones zpu
    	on t."PULocationID" = zpu."LocationID"
    join zones zdo
    	on t."DOLocationID" = zdo."LocationID"	
limit 100;
```

Groupyby

```python
%%sql

SELECT 
    tpep_pickup_datetime,
    tpep_dropoff_datetime,
    tolls_amount ,
    DATE_TRUNC('DAY', tpep_dropoff_datetime)
FROM
    yellow_taxi_data t LEFT JOIN zones zpu
    	on t."PULocationID" = zpu."LocationID"
    join zones zdo
    	on t."DOLocationID" = zdo."LocationID"	
limit 100;
```

```python
%%sql

SELECT 
    CAST(tpep_dropoff_datetime AS DATE) as "day",
    COUNT(1)
FROM
    yellow_taxi_data t
GROUP BY
    CAST(tpep_dropoff_datetime AS DATE);
```

Other kinds of aggregation

```python
%%sql

SELECT
    COUNT(1) as "count",
    MAX(tolls_amount),
    MAX(passenger_count)
FROM 
    yellow_taxi_data
```

```python
%%sql

SELECT
    CAST(tpep_dropoff_datetime AS DATE) as "Day",
    COUNT(1) as "count",
    MAX(tolls_amount),
    MAX(passenger_count)
FROM
    yellow_Taxi_data t
GROUP BY
    CAST(tpep_dropoff_datetime as DATE)
ORDER BY
    "count" DESC;
```

```python
%%sql

SELECT
    CAST(tpep_dropoff_datetime AS DATE) as "day",
    "DOLocationID",
    COUNT(1) as "count",
    MAX(tolls_amount),
    MAX(passenger_count)
FROM
    yellow_Taxi_data t
GROUP BY
    1,2
ORDER BY
    "day" DESC, 
    "DOLocationID" ASC;
```
